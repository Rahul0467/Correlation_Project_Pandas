```python
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use("ggplot")
from matplotlib import figure

%matplotlib inline
matplotlib.rcParams["figure.figsize"] = (12, 8) 
```

```python
# Read the data

df = pd.read_csv("movies.csv")
df.head()
```

```python
# Copy the dataset 

df2 = df.copy()
```

```python
# Looking for missing data

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print(("{} - {}%").format(col, pct_missing))
```

```python
# Replacing na values and changing the datatypes

df["budget"].fillna(0, inplace= True)
df["budget"] = df["budget"].astype("int64")

df["gross"].fillna(0, inplace= True)
df["gross"] = df["gross"].astype("int64") 

df["votes"].fillna(0, inplace= True)
df["votes"] = df["votes"].astype("int64")

df["runtime"].fillna(0, inplace= True)
df["runtime"] = df["runtime"].astype("int64") 
df.head()
```

```python
# Sorting the data

df.sort_values(by= ["gross"], inplace=True, ascending=False)
df.head()
```

```python
pd.set_option("display.max_rows", None)
```

```python
# Dropping duplicates

df.drop_duplicates().head()
```

```python
# Finding correlation
# Predictions
# The budget and gross have a positive correlation
# The Production company will have a high correlation
# Votes have a positive correlation
```

```python
# Scatter plot with budget vs gross

plt.scatter(x=df["budget"], y=df["gross"])
plt.title("Budget vs Gross")
plt.xlabel("Budget of the film")
plt.ylabel("Gross Earnings")
plt.show()
```

```python
# Budget vs Gross using seaborn

sns.regplot(x="budget", y="gross", data=df, scatter_kws={"color":"black"}, line_kws={"color":"purple"})
plt.title("Budget vs Gross")
plt.xlabel("Budget of the film")
plt.ylabel("Gross Earnings")
```

```python
df.corr(method="pearson")
```

```python
corr_matrix= df.corr(method="pearson")
sns.heatmap(corr_matrix, annot=True)
plt.title("Correlation between Numeric Features")
plt.xlabel("Movie Features")
plt.ylabel("Movie Features")
plt.show()
```

```python
# The highest correlation is between the budget of the film and gross earnings followed by votes and gross earnings
```

```python
df_num = df2

for col_name in df_num.columns:
    if(df_num[col_name].dtype == "object"):
        df_num[col_name] = df_num[col_name].astype("category")
        df_num[col_name] = df_num[col_name].cat.codes
        
df_num.sort_values(by="gross", inplace=True, ascending=False)
df_num.head()
```

```python
df.head()
```

```python
corr_mat= df_num.corr(method="pearson")
sns.heatmap(corr_mat, annot=True)
plt.title("Correlation between variables")
plt.xlabel("Movie Features")
plt.ylabel("Movie Features")
plt.show()
```

```python
correlation_mat = df_num.corr(method="pearson")
corr_pairs = correlation_mat.unstack()
sorted_pairs = corr_pairs.sort_values()
high_corr = sorted_pairs[(sorted_pairs) > 0.5]
high_corr
```
